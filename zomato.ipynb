```python
rating = df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'count'})
```

```python
rating
```

## Observation 

1. When rating between 4.5 to 4.9 => 'Excellent'
2. When rating between 4.0 to 4.4 => 'Very Good'
3. When rating between 3.5 to 3.9 => 'Good'
4. When rating between 2.5 to 3.4 => 'Average'
5. When rating between 1.8 to 2.4 => 'Poor'
6. When rating 0.00 => 'Not rated'

```python
plt.figure(figsize=(12,8))
sns.barplot(data=rating, x = 'Aggregate rating', y='count',hue='Rating color', palette=['blue','red','orange','yellow','green','green'])
```

## Observation 
1. Not rated count is very high
2. Most of the ratings is between 2.7 to 4.2
